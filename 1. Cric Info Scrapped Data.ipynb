```python
import requests
from bs4 import BeautifulSoup
```

```python
base_site="https://www.espncricinfo.com/rankings/content/page/211271.html"
```

```python
response=requests.get(base_site)
```

```python
response.ok
```

```python
response.status_code
```

```python
html=response.text
```

```python
read=BeautifulSoup(html,"html.parser")
```

```python
read
```

```python
all_tables=read.find_all("table",class_="StoryengineTable")
```

```python
odis=all_tables[1]
```

```python
odis.caption.text
```

```python
print(odis.text)
```

```python
col=[]
for i in odis.find_all("th"):
    print(i.text)
    col.append(i.text)
```

```python
col
```

```python
table=odis.find_all("td")
```

```python
table_list=[]
start=0
end=5
for i in range(20):
    Pos, Team, Matches, Points, Rating =table[start:end]
    start+=5
    end+=5
    table_list.append([Pos.text, Team.text, Matches.text, Points.text, Rating.text])
   
```

```python
table_list
```

```python
import pandas as pd
odi_ranking=pd.DataFrame(table_list , columns=col)
```

```python
odi_ranking
```

```python
odi_ranking.to_csv("odi.csv",index=False)
```
